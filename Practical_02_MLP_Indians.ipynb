```python
from keras.models import Model, Sequential
from keras.layers import Input, Dense

import numpy
#fix random seed if you want reproducibility
numpy.random.seed(7)
#after you run the code, try it with another random seed. What do you notice?
```

```python
# load data directly in your computer
dataset = numpy.loadtxt("indians.cvs", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
#for every function used after this point, check the keras documentation
```

```python
X
```

```python
Y
```

```python
#creating the network
model = Sequential() # Initialize a sequential model (layers are added one after another in a linear stack)
model.add(Dense(12, input_dim=8, activation='relu')) #  Add the first hidden layer # 12 neurons (nodes)
model.add(Dense(8, activation='relu')) # Add a second hidden layer
model.add(Dense(1, activation='sigmoid')) # Add the output layer
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the model
model.fit(X, Y, epochs=150, batch_size=10)
```

```python
# evaluate the model
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# calculate predictions
predictions = model.predict(X)
# round predictions
rounded = [round(x[0]) for x in predictions]
print(rounded)
```
