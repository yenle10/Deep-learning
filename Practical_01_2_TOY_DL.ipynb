```python
# run all as a whole and comments added 

# from tensorflow.keras.models import Sequential
# from tensorflow.keras.layers import Dense
# from tensorflow.keras.optimizers import SGD
# import numpy as np

# Create the optimizer
optimizer = SGD(learning_rate=0.5)
# SGD = Stochastic Gradient Descent
# learning_rate=0.5 is quite high (usually we use 0.01 or 0.001)

# Initialize the neural network model
model = Sequential()                                 # Creates an empty sequential model (layers added one after another)

# Add layers

model.add(Dense(2, input_dim=2, activation='relu'))  # First hidden layer:
# Dense = a fully connected layer → every neuron connects to all inputs
# 2 neurons → like 2 circles (z1, z2)
# input_dim=2 → each neuron receives 2 inputs (x1, x2)
# Each neuron computes: z = w0 + w1*x1 + w2*x2, then applies ReLU

model.add(Dense(2, activation='sigmoid'))            # Output layer:
# Each neuron takes outputs from previous layer (z1, z2)
# Computes weighted sum again, then applies sigmoid (output between 0 and 1)

# Compile the model (configure how it will learn)
model.compile(loss='mean_squared_error', optimizer=optimizer)
# loss = MSE (Mean Squared Error) 
# optimizer = SGD: updates neuron weights using gradient descent

model.summary()                                      # Prints the structure of the model (layers, parameters)

# ====================== Print initial weights ======================
print("Weights and biases of the layers before training the model: \n")

for layer in model.layers:
    print(layer.name)
    print("Weights")
    print("Shape: ", layer.get_weights()[0].shape, '\n', layer.get_weights()[0])
    print("Bias")
    print("Shape: ", layer.get_weights()[1].shape, '\n', layer.get_weights()[1], '\n')
# Keras initializes automatically weights and bias when the layer is built

# ====================== Prepare training data ======================
inp = np.array([[1,2], [1,2], [1,2]])   # Input data: 3 samples, each with 2 features
trg = np.array([[0.1,0.7], [0.1,0.7], [0.1,0.7]])  # Target data: 3 samples, each with 2 target values

# Check shapes (good practice)
inp.shape   # Should be (3, 2)
trg.shape   # Should be (3, 2)

# ====================== First training ======================
model.fit(inp, trg, epochs=1000, batch_size=3, verbose=1)
# Trains the model for 1000 epochs
# batch_size=3 → uses all 3 samples at once (full batch gradient descent)
# verbose=1 → shows progress bar + loss per epoch

# ====================== Manually set weights ======================
# Setting custom weights for layer 0 (first hidden layer)
l = []
x = np.array([[-1, -0.5], [1, 2]])      # Weight matrix (2 inputs → 2 neurons)
y = np.array([2.1, -4])                 # Bias vector
l.append(x)
l.append(y)
model.layers[0].set_weights(l)          # Override weights & biases of first layer

# Setting custom weights for layer 1 (output layer)
l = []
x = np.array([[2, -0.4], [-1, 1.5]])    # Weight matrix (2 hidden neurons → 2 outputs)
y = np.array([-0.3, 0.6])               # Bias vector
l.append(x)
l.append(y)
model.layers[1].set_weights(l)          # Override weights & biases of output layer

# ====================== Print weights again ======================
print("Weights and biases of the layers before training the model: \n")
for layer in model.layers:
    print(layer.name)
    print("Weights")
    print("Shape: ", layer.get_weights()[0].shape, '\n', layer.get_weights()[0])
    print("Bias")
    print("Shape: ", layer.get_weights()[1].shape, '\n', layer.get_weights()[1], '\n')

# ====================== Train again with new weights ======================
model.fit(inp, trg, epochs=1000, batch_size=3, verbose=1)
# Trains the model again starting from the manually set weights
```

### Reason for each step:

1. **Print initial weights** (before any training)  
   → To **show the random weights** that Keras automatically creates when the model is built.  
   This is for **learning/education** — so we can see what the starting point looks like.

2. **First training** (`model.fit(...)` the first time)  
   → Lets the model **learn normally** from random weights using SGD for 1000 epochs.  
   This demonstrates standard training behavior.

3. **Manually set weights** (`set_weights()`)  
   → **Overrides** the learned (or random) weights with **specific values** chosen by the programmer.  
   This is **not normal** in real projects. It's done here for **experimentation/teaching** — to test how the model behaves when starting from known/specific weights instead of random ones.

4. **Print weights again** (after manual setting)  
   → To **verify** that the manual weights were correctly applied.  
   Good practice when we change weights manually.

5. **Train again** (second `model.fit()`)  
   → Continues training **starting from the manually set weights**.  
   This lets we observe how the network updates from our chosen starting point (instead of random).

### Is this necessary for a normal Sequential model workflow?

**No.**  
This is **not** a standard workflow.

In normal deep learning, usually do only this:
- Build model → Compile → `model.fit()` once (or with callbacks)

The extra steps (print weights + manual `set_weights` + train twice) are **only for educational or debugging purposes**.  
They are used for purposes:
- Understand how weights change during training
- Compare training from random vs specific weights
- Demonstrate concepts in a classroom/tutorial

**In short**:  
The first train + manual set + second train is **not required**. It’s added intentionally to **compare** normal training vs training from custom starting weights. It’s a teaching/demo trick, not a necessary part of using a Sequential model.
